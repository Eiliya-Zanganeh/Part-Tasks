```python
import stanza
```

```python
nlp = stanza.Pipeline(lang='fa', processors='tokenize')
doc = nlp('من در حال درس خواندن هستم')
for i, sentence in enumerate(doc.sentences):
    print(f'====== Sentence {i+1} tokens =======')
    for token in sentence.tokens:
      print(f'id: {token.id}\t', end='')
      print(f'text: {token.text}')
```

```python
nlp = stanza.Pipeline(lang='fa', processors='tokenize,mwt,pos,lemma,depparse')
doc = nlp('من در حال درس خواندن هستم')

for sent in doc.sentences:
  for word in sent.words:
    print(f'id: {word.id}\t', end='')
    print(f'word: {word.text}\t', end='')
    print(f'head id: {word.head}\t', end='')
    if word.head > 0:
      print(f'head: {sent.words[word.head-1].text}\t', end='')
    else:
      print("head: root\t", end='')
    print(f'deprel: {word.deprel}')
```
